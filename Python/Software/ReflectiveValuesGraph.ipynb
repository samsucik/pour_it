```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open("data.txt", "r")
dictx = {}
dicty = {}
name = ''
for line in file:
    args = line.split()
    if(args[1] == 'degrees'):
        name = args[0] + ' ' + args[1]
        dictx[name] = []
        dicty[name] = []
    else:
        dictx[name].append(float(args[0]))
        dicty[name].append(int(args[1]))
```

```python
angles = ['90 degrees', '67.5 degrees', '45 degrees', '22.5 degrees']
for angle in angles:
    plt.title(angle + '; duty_cycle_sp = 30; lighting conditions = dark')
    plt.xlabel('Time (in seconds)')
    plt.ylabel('Reflective value read')
    plt.plot(np.array(dictx[angle]) - dictx[angle][0], np.array(dicty[angle]))
    plt.show()
    plt.close()
    
```
